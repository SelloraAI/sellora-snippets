```python
# üìÅ Upload the CSV from Day 2
uploaded = files.upload()

# üìÑ Load the product data
df = pd.read_csv("sellora_products_with_embeddings.csv")

# üîÅ Convert embeddings from text to real list
df["embedding"] = df["embedding"].apply(ast.literal_eval)

print("‚úÖ File loaded and parsed!")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# üîê Paste your real API key again here (Colab forgets it when restarted)
openai.api_key = "sk-proj" <- YOUR API KEY HERE

def search_products(query, df, top_n=3):
    # üß† Generate embedding for the search query
    query_emb = openai.embeddings.create(
        input=query,
        model="text-embedding-3-small"
    ).data[0].embedding

    # üîç Compare query to every product embedding
    sims = cosine_similarity([query_emb], df["embedding"].tolist())[0]

    # üìä Add similarity scores to dataframe
    df["similarity"] = sims

    # üìà Sort and return top N results
    return df.sort_values("similarity", ascending=False).head(top_n)[
        ["Product Name", "Description", "similarity"]
    ]
```

```python
search_products("wireless headphones with long battery life", df)
```
